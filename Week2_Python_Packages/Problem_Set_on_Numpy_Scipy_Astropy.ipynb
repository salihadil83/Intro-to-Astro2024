Write a Simple python function to calculate **Escape Velocity** of any celestrial body. Calculate the value for earth.

### Hints

It is the minimum velocity needed to escape from gravitational binding.

```python
import math
def calculate_escape_velocity(mass, radius):

    G = 6.67 * 10**-11  # Evrensel Gravitasyon Sabiti
    v_e = math.sqrt(2 * G * mass / radius)
    return v_e
# Dünya için kaçış hızını hesaplama
earth_mass = 5.97 * 10**24  # Dünya'nın kütlesi (kg)
earth_radius = 6.371 * 10**6  # Dünya'nın yarıçapı (m)
earth_escape_velocity = calculate_escape_velocity(earth_mass, earth_radius)
print(f"Dünya'nın kaçış hızı: {earth_escape_velocity:.2f} m/s")
```

## Question-2

Write a function which can calculate the value of $\pi$ using random numbers using **numpy** library. Try using 1000 sample

```python
import numpy as np
```

```python
def calculate_o2_rms_velocity(num_samples=1000):
    m_o2 = 32.0 * 1.66e-27  # Oksijen molekülünün kütlesi (kg)
    T = 300  # Sıcaklık (K)
    k = 1.38e-23  # Boltzmann sabiti (J/K)
    masses = np.random.normal(m_o2, m_o2 * 0.01, num_samples)
    v_rms_o2s = np.sqrt(3 * k * T / masses)
    avg_v_rms_o2 = np.mean(v_rms_o2s)
    return avg_v_rms_o2
rms_velocity = calculate_o2_rms_velocity(num_samples=1000)
print(f"Oksijen molekülünün ortalama rms hızı: {rms_velocity:.2f} m/s")
```

## Question-5

Use Newton's Gravitational Law along with Newton's 2nd law of motion to write the differential equation which earth will follow due to Sun's gravity (sun is fixed in it's place).

1. Now use Scipy to solve the equation. The constants needed must be imported from Astropy.
2. Plot your solution , i.e., x-y plot.

```python
import numpy as np
from scipy.integrate import solve_ivp
from astropy.constants import G, M_sun, M_earth

# Sabitler
G = G.value  # Gravitasyon sabiti
m_s = M_sun.value  # Güneş'in kütlesi
m_d = M_earth.value  # Dünya'nın kütlesi

# Başlangıç koşulları
r0 = 1.496e11  # Güneş-Dünya arası ortalama mesafe (m)
v0 = 2.9783e4  # Dünya'nın Güneş etrafındaki hızı (m/s)

# Diferansiyel denklemin fonksiyonu
def dxdt(t, x):
    r = np.sqrt(x[0]**2 + x[1]**2)
    return [x[2], x[3], -G * m_s * x[0] / r**3, -G * m_s * x[1] / r**3]

# Diferansiyel denklemi çöz
t_span = (0, 3.154e7)  # 1 yıllık zaman aralığı (s)
y0 = [r0, 0, 0, v0]  # Başlangıç koşulları
sol = solve_ivp(dxdt, t_span, y0, method='RK45')

# Çözümü çiz
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.plot(sol.y[0], sol.y[1])
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Dünya\'nın Güneş Etrafındaki Hareketi')
plt.axis('equal')
plt.show()
```

## Question-6

**NASA Cosmic Background Explorer** (COBE) satellite carried an instrument, **FIRAS** (Far-Infrared Absolute Spectrophotometer) to measure the cosmic microwave background (CMB) radiation, which was confirmed to be distributed according to a black-body curve in accordance with the big bang theory:
$$ I(\nu, T) = \frac{2 h \nu^3 c^2}{\exp(\frac{hc\nu}{k_{b}T})-1}$$
where where the radiation frequency is expressed in wavenumbers, $cm^{-1}$, and the speed of light, c is taken to be in $cm-s^{-1}$.

The data file is cmb_data.txt, which contains measured $I(\nu)$ based on the FIRAS observations. Use scipy `curve_fit` to determine $T$, i.e., the Temperature parameter, along with error.

Note: In the file $I$ is in $erg\cdot s^{-1}\cdot cm^{-1}\cdot sr^{-1}$. Take the estimated $\sigma$ error in the measurement to be $2\times 10^{-6}$$erg\cdot s^{-1}\cdot cm^{-1}\cdot sr^{-1}$.

```python
#When error is known the way to use curve_fit is 
#para, pcov = curve_fit(x, y, fit_func, p0=(T0,), sigma=σ, absolute_sigma=True)
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Veri dosyasını yükle
data = np.loadtxt('cmb_data.txt')
wavenumber = data[:, 0]
intensity = data[:, 1]
err = 2e-6  # Ölçüm hatası
```

```python
def planck(nu, T):
    """Planck ışınım yasasına göre ışınım yoğunluğu"""
    h = 6.626e-34  # Planck sabiti
    k = 1.381e-23  # Boltzmann sabiti
    c = 3e8  # Işık hızı
    return 2 * h * nu**3 / (c**2 * (np.exp(h * nu / (k * T)) - 1))
```

```python
popt, pcov = curve_fit(planck, wavenumber, intensity, sigma=err*np.ones_like(intensity), p0=2.725)

T = popt[0]
T_err = np.sqrt(pcov[0,0])

print(f"Sıcaklık parametresi: {T:.3f} ± {T_err:.3f} K")
```

```python
# Planck dağılımı çizimi
plt.figure(figsize=(8, 6))
plt.errorbar(wavenumber, intensity, yerr=err*np.ones_like(intensity), fmt='o', label='Gözlenen veri')
plt.plot(wavenumber, planck(wavenumber, T), 'r-', label=f'Planck dağılımı (T={T:.3f} ± {T_err:.3f} K)')
plt.xlabel('Dalga sayısı (cm$^{-1}$)')
plt.ylabel('Işınım yoğunluğu ($erg \cdot s^{-1} \cdot cm^{-1} \cdot sr^{-1}$)')
plt.title('Kozmik Mikrodalga Arka Plan Radyasyonu')
plt.legend()
plt.grid()
plt.show()
```

First, we need the mass of the proton. The mass of a proton is approximately:
m_proton = 1.6726 × 10^-27 kg

The speed of light, c = 3 × 10^8 m/s

In joules, the rest mass energy of the proton is:
E = m_proton * c^2
E = (1.6726 × 10^-27 kg) * (3 × 10^8 m/s)^2
E = 1.5033 × 10^-10 J

Now let's calculate it in MeV (Mega electron volts). 1 MeV = 1.6022 × 10^-13 J

E (MeV) = E (J) / (1.6022 × 10^-13 J/MeV)
E (MeV) = (1.5033 × 10^-10 J) / (1.6022 × 10^-13 J/MeV)
E (MeV) = 938.27 MeV

In summary:

The rest mass energy of a proton is 1.5033 × 10^-10 J.
The rest mass energy of a proton is 938.27 MeV.
Therefore, the rest mass energy of a proton has been calculated in both joules and MeV.

### Theory needed for problem-8

**Diurnal motion** is an astronomical term referring to the apparent motion of celestial objects (e.g. the Sun and stars) around Earth, or more precisely around the two celestial poles, over the course of one day.

It is caused by Earth's rotation around its axis, so almost every star appears to follow a circular arc path, called the diurnal circle,often depicted in star trail photography.

So, From the viewpoint of an observer on Earth, the apparent motion of an object on the celestrial sphere follows an arc above the horizon, which is called **diurnal arc**.

![Diurnal arc](https://webapps-cdn.esri.com/CDN/gis-dictionary/Media/diurnal-arc.gif)

Here, Diurnal Arc of a Star moving around the celestrial sphere(red line) in the horizontal system of an observer at lattitude $\phi$. Since, the equatorial plane is inclined by the angle $90^{\circ} - \phi$ against the horizontal plane, the upper culmination of the star at the meridian is given by $a_{max} = 90^{\circ} - \phi + \delta$, where $\delta$ is the declination. The star **rises** at hour angle $h_{rise}$, reaches it's highest altitude when it crosses the meridian at $h=0$ and sets at the horizon at $h_{set} = -h_{rise}$. The value can be given by,
$$ \cos(h_{rise}) = -\tan(\delta) \tan(\phi) $$

**Sidereal Time** is the time for which the star is visible on sky. It is given by $T = 2h_{set}$.

**Problem-8**: Find how long the star `Betelgeuse` is present on sky from my location (Jadavpur, Kolkata, India).

#Solution
* Location: Jadavpur, Kolkata, India
* Star: Betelgeuse
To calculate the sidereal time for Betelgeuse at this location, we need to find the hour angle at which Betelgeuse rises above the horizon.

The formula to calculate the hour angle at rise (h_rise) is:

cos(h_rise) = -tan(δ) tan(φ)

Where:

* δ is the declination of Betelgeuse
* φ is the latitude of the observer's location
Let's look up the values:

* Declination of Betelgeuse (δ) = 7°24'
* Latitude of Jadavpur, Kolkata (φ) = 22°32'N
Plugging these values into the formula:
cos(h_rise) = -tan(7°24') tan(22°32')
h_rise = 4.2 hours

Therefore, the sidereal time for Betelgeuse, or the duration for which Betelgeuse is present in the sky at the Jadavpur, Kolkata location, is:

Sidereal Time (T) = 2 * h_rise
T = 2 * 4.2 hours
T = 8.4 hours

So the star Betelgeuse is present in the sky for approximately 8.4 hours at the Jadavpur, Kolkata location.

## Question-9

The **Declination** of sun $\delta_s$ is given by,
$$ \delta_s = -\arcsin\Bigg(\sin(\epsilon_0)\cos\Big(\frac{360}{365.24}(N+10)\Big)\Bigg)$$
where $\epsilon_0 = 23.44^{\circ}$ and $N$ is the difference in days starting from 1st january.

Make a plot of how the length of day changes over the year in your location.

```python
# Constants
epsilon_0 = 23.44  # Obliquity of Earth's axis
latitude = 41.01   # Latitude of Istanbul, Turkey in degrees

# Calculate declination of Sun for each day of the year
days = np.arange(1, 366)
declination = -np.arcsin(np.sin(np.radians(epsilon_0)) * np.cos(np.radians(360/365.24 * (days + 10))))

# Calculate hour angle at sunrise/sunset for each day
h_rise = np.arccos(-np.tan(np.radians(declination)) * np.tan(np.radians(latitude)))
h_set = -h_rise

# Calculate day length for each day
day_length = 2 * (h_set - h_rise)
day_length[day_length < 0] = 0  # Avoid negative day lengths near solstices

# Plot the day length over the year
plt.figure(figsize=(12, 6))
plt.plot(days, np.degrees(day_length))
plt.xlabel('Day of the Year')
plt.ylabel('Day Length (hours)')
plt.title('Variation of Day Length over the Year in Istanbul, Turkey')
plt.grid()
plt.show()
```
